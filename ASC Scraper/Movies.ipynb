```python
for i in range(3,4):
        while True :
                try:
                        title=driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[2]/div[1]/a/h3")
                        Movies.iloc[(i-1),0]=title.text

                        image=driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[1]/div/div[2]/img")
                        Movies.iloc[(i-1),1]=image.get_attribute('src')

                        for j in range(2,4):
                                try:
                                        elem= driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[2]/div[2]/span[{j}]")
                                        Movies.iloc[(i-1),(j+1)]=elem.text
                                except:
                                        continue
                        Rating=driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[2]/span/div/span/span[1]")
                        number=driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[2]/span/div/span/span[2]")
                        Movies.iloc[(i-1),5]=Rating.text+number.text

                        plot= driver.find_element(By.XPATH, f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[2]/div/div")
                        Movies.iloc[(i-1),10]=plot.text

                        info=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li[{i}]/div/div/div/div[1]/div[2]/div[1]/a")
                        info.click()
                        scroll_length = 4100
                        driver.execute_script(f"window.scrollBy(0, {scroll_length});")

                        Directors=""
                        for j in range(1,10):
                                try:
                                        Director=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[2]/div/ul/li[1]/div/ul/li[{j}]/a")
                                        #                                       /html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[2]/div/ul/li[1]/div/ul/li/a
                                        Directors=Directors+Director.text+", "
                                except:
                                        Movies.iloc[(i-1),9]=Directors[:-2]
                                        break

                        for k in range (5,8):
                                try:
                                        Genres=""
                                        for j in range(1,10):
                                                try:
                                                        Genre=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[{k}]/div[2]/ul[2]/li[2]/div/ul/li[{j}]/a")
                                                        #                                    /html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[6]/div[2]/ul[2]/li[2]/div/ul/li[1]/a
                                                        Genres=Genres+Genre.text+", "
                                                except:
                                                        Movies.iloc[(i-1),7]=Genres[:-2]
                                                        break
                                except:
                                        continue
                        scroll_length = 3000
                        for k in range (10,14):
                                try:
                                        driver.execute_script(f"window.scrollBy(0, {scroll_length});")
                                        
                                        Release_Date=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[{k}]/div[2]/ul/li[1]/div/ul/li/a")
                                        Movies.iloc[(i-1),2]=Release_Date.text

                                        Countries=""
                                        for j in range(1,10):
                                                try:
                                                        Country=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[{k}]/div[2]/ul/li[2]/div/ul/li[{j}]/a")
                                                        Countries=Countries+Country.text+", "
                                                except:
                                                        Movies.iloc[(i-1),6]=Countries[:-2]
                                                        break

                                        Languages=""
                                        for j in range(1,10):
                                                try:
                                                        Language=driver.find_element(By.XPATH,f"/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[{k}]/div[2]/ul/li[4]/div/ul/li[{j}]/a")
                                                        Languages=Languages+Language.text+", "
                                                except:
                                                        Movies.iloc[(i-1),8]=Languages[:-2]
                                                        break  
                                        
                                        break     
                                except:
                                        continue
                        driver.back()
                        break
                except:
                        driver.implicitly_wait(2)
                        scroll_length = 300
                        driver.execute_script(f"window.scrollBy(0, {scroll_length});")
                        continue
        
        
```
